```python
#!/usr/bin/env python3
import torch
import sys
import platform
import subprocess
import os

def get_cpu_info():
    """Linux環境でCPUモデル名を取得する"""
    try:
        # /proc/cpuinfo から model name を取得
        with open("/proc/cpuinfo", "r") as f:
            for line in f:
                if "model name" in line:
                    return line.split(":")[1].strip()
    except:
        pass
    # 取得できなかった場合は簡易情報
    return platform.processor()

def get_ram_info():
    """Linux環境でメモリ容量を取得する"""
    try:
        with open("/proc/meminfo", "r") as f:
            for line in f:
                if "MemTotal" in line:
                    kb = int(line.split(":")[1].strip().split()[0])
                    gb = kb / (1024 ** 2)
                    return f"{gb:.1f} GB"
    except:
        return "Unknown"

def main():
    print("="*40)
    print(" Experimental Environment Specs")
    print("="*40)

    # 1. GPU Info
    print(f"[GPU]")
    if torch.cuda.is_available():
        device_count = torch.cuda.device_count()
        print(f"  Count        : {device_count}")
        for i in range(device_count):
            name = torch.cuda.get_device_name(i)
            # VRAM容量 (Byte -> GB)
            vram = torch.cuda.get_device_properties(i).total_memory / (1024**3)
            print(f"  Model ({i})    : {name}")
            print(f"  VRAM ({i})     : {vram:.1f} GB")
        print(f"  CUDA Version : {torch.version.cuda}")
        # CuDNN version
        if torch.backends.cudnn.is_available():
            print(f"  CuDNN Version: {torch.backends.cudnn.version()}")
    else:
        print("  No GPU detected (Using CPU)")

    print("-" * 40)

    # 2. CPU Info
    print(f"[CPU]")
    print(f"  Model        : {get_cpu_info()}")
    print(f"  Architecture : {platform.machine()}")
    # 論理コア数
    print(f"  Logical Cores: {os.cpu_count()}")

    print("-" * 40)

    # 3. RAM Info
    print(f"[Memory]")
    print(f"  System RAM   : {get_ram_info()}")

    print("-" * 40)

    # 4. Software Info
    print(f"[Software]")
    print(f"  OS           : {platform.system()} {platform.release()}")
    print(f"  Python       : {sys.version.split()[0]}")
    print(f"  PyTorch      : {torch.__version__}")

    print("="*40)

if __name__ == "__main__":
    main()
```
